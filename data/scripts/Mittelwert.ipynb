```python
participant = glob.glob('.../data_behav/*.csv')
```

```python
#create lists to append values and to store them in the dataframe df with mean reactions
list_accuracy_unconscious =[]
list_accuracy_conscious=[]

list_accuracy_happy =[]
list_accuracy_sad=[]
list_accuracy_neutral =[]

list_accuracy_unconscious_happy=[]
list_accuracy_unconscious_sad =[]
list_accuracy_unconscious_neutral=[]

list_accuracy_conscious_happy=[]
list_accuracy_conscious_sad =[]
list_accuracy_conscious_neutral =[]

participant_ID_list = []
```

```python
for i in participant:
    #read in csv for every participant
    participantcsv = pd.read_csv(i, sep =",") #\t , ; ""
    
    #Delete 5 example trials
    participantcsv=participantcsv.drop([2,3,4,5,6])
    
    #get correct trials
    def correctness(participantcsv):
        if participantcsv['corrAnsPrimeStudy'] == participantcsv['key_resp_study.keys']:
            return 1
        else:
            return 0
    participantcsv['correct'] = participantcsv.apply(correctness, axis=1) 
    
    
    #Just get values smaller than 1.5
    participantcsv = participantcsv[(participantcsv['key_resp_study.rt'] <= 1.5)] 
    #Drop trials without reaction
    participantcsv = participantcsv.dropna(subset=['key_resp_study.rt'])

    #Get correct answers
    corr_answers = participantcsv['correct']
    reaction_time = participantcsv['key_resp_study.rt']

    #Create lists with trials per condition
    happy = participantcsv[(participantcsv['emotionPrimeStudy'] =='happy')]
    sad = participantcsv[(participantcsv['emotionPrimeStudy'] =='sad')]
    neutral = participantcsv[(participantcsv['emotionPrimeStudy'] =='neutral')]
    
    unconscious = participantcsv[(participantcsv['study mask time'] == 1) | (participantcsv['study mask time'] == 2) | (participantcsv['study mask time'] == 3)] 
    conscious = participantcsv[(participantcsv['study mask time'] == 17)] 
    
    unconscious_happy = unconscious[(unconscious['emotionPrimeStudy'] == 'happy')]
    unconscious_sad = unconscious[(unconscious['emotionPrimeStudy'] == 'sad')]
    unconscious_neutral = unconscious[(unconscious['emotionPrimeStudy'] == 'neutral')]
                                      
    conscious_happy = conscious[(conscious['emotionPrimeStudy'] == 'happy')]
    conscious_sad = conscious[(conscious['emotionPrimeStudy'] == 'sad')]
    conscious_neutral = conscious[(conscious['emotionPrimeStudy'] == 'neutral')]
    
    
    #Excecute functions and get one value for each participant
    #accuracy
    val_unconscious = accuracy_unconscious(participantcsv,i)
    val_conscious = accuracy_conscious(participantcsv,i)
    
    val_happy = accuracy_happy(participantcsv,i)
    val_sad = accuracy_sad(participantcsv,i)
    val_neutral = accuracy_neutral(participantcsv,i)                
                                    
    val_unconscious_happy = accuracy_unconscious_happy(participantcsv,i)
    val_unconscious_sad = accuracy_unconscious_sad(participantcsv,i)
    val_unconscious_neutral = accuracy_unconscious_neutral(participantcsv,i)
    
    val_conscious_happy = accuracy_conscious_happy(participantcsv,i)
    val_conscious_sad = accuracy_conscious_sad(participantcsv,i)
    val_conscious_neutral = accuracy_conscious_neutral(participantcsv,i)
    
    #rt
    val_rt_unconscious = rt_unconscious(participantcsv,i)
    val_rt_conscious = rt_conscious(participantcsv,i)

    val_rt_happy = rt_happy(participantcsv,i)
    val_rt_sad = rt_sad(participantcsv,i)
    val_rt_neutral = rt_neutral(participantcsv,i)                
                            
    val_rt_unconscious_happy = rt_unconscious_happy(participantcsv,i)
    val_rt_unconscious_sad = rt_unconscious_sad(participantcsv,i)
    val_rt_unconscious_neutral = rt_unconscious_neutral(participantcsv,i)

    val_rt_conscious_happy = rt_conscious_happy(participantcsv,i)
    val_rt_conscious_sad = rt_conscious_sad(participantcsv,i)
    val_rt_conscious_neutral = rt_conscious_neutral(participantcsv,i)
    
    participant_ID = name(participantcsv,i)

    
    #append values to append individual value to one list                             
    list_accuracy_unconscious.append(val_unconscious)
    list_accuracy_conscious.append(val_conscious)
    
    list_accuracy_happy.append(val_happy)
    list_accuracy_sad.append(val_sad)
    list_accuracy_neutral.append(val_neutral)
    
    list_accuracy_unconscious_happy.append(val_unconscious_happy)
    list_accuracy_unconscious_sad.append(val_unconscious_sad)
    list_accuracy_unconscious_neutral.append(val_unconscious_neutral)
    
    list_accuracy_conscious_happy.append(val_conscious_happy)
    list_accuracy_conscious_sad.append(val_conscious_sad)
    list_accuracy_conscious_neutral.append(val_conscious_neutral) 
    
    participant_ID_list.append(participant_ID)

    list_rt_unconscious.append(val_rt_unconscious)
    list_rt_conscious.append(val_rt_conscious)
    
    list_rt_happy.append(val_rt_happy)
    list_rt_sad.append(val_rt_sad)
    list_rt_neutral.append(val_rt_neutral)
    
    list_rt_unconscious_happy.append(val_rt_unconscious_happy)
    list_rt_unconscious_sad.append(val_rt_unconscious_sad)
    list_rt_unconscious_neutral.append(val_rt_unconscious_neutral)
    
    list_rt_conscious_happy.append(val_rt_conscious_happy)
    list_rt_conscious_sad.append(val_rt_conscious_sad)
    list_rt_conscious_neutral.append(val_rt_conscious_neutral) 
    
    study_number = 'study1'
    
    #Create dataframe 
    accuracy_data = {'name': participant_ID_list,
                     
                     'accuracy happy':list_accuracy_happy,
                     'accuracy sad':list_accuracy_sad,
                     'accuracy neutral':list_accuracy_neutral,
                     'accuracy unconscious':list_accuracy_unconscious,
                     'accuracy conscious':list_accuracy_conscious,
                     
                     'accuracy unconscious happy':list_accuracy_unconscious_happy,
                     'accuracy unconscious sad':list_accuracy_unconscious_sad,
                     'accuracy unconscious neutral':list_accuracy_unconscious_neutral,
                     
                     'accuracy conscious happy':list_accuracy_conscious_happy,
                     'accuracy conscious sad':list_accuracy_conscious_sad,
                     'accuracy conscious neutral':list_accuracy_conscious_neutral,
                     
                     'rt happy':list_rt_happy,
                     'rt sad':list_rt_sad,
                     'rt neutral':list_rt_neutral,
                     'rt unconscious':list_rt_unconscious,
                     'rt conscious':list_rt_conscious,
                     
                     'rt unconscious happy':list_rt_unconscious_happy,
                     'rt unconscious sad':list_rt_unconscious_sad,
                     'rt unconscious neutral':list_rt_unconscious_neutral,
                     
                     'rt conscious happy':list_rt_conscious_happy,
                     'rt conscious sad':list_rt_conscious_sad,
                     'rt conscious neutral':list_rt_conscious_neutral,
                     
                    'study_number':study_number}
                                    
    mean_accuracy_1 = pd.DataFrame(data=accuracy_data).round(4)
    mean_accuracy_1['name'] = mean_accuracy_1['name'].replace(['0HZ'],'005')
```
