```python
#read in csv data generated by PsychoPy
participant = glob.glob('.../data_behav_2/*.csv')
participant
```

```python
#create lists to append values and to store them in the dataframe df with mean reactions
list_accuracy_uncon_8ms_happy =[]
list_accuracy_uncon_8ms_sad =[]
list_accuracy_uncon_8ms_neutral =[]

list_accuracy_uncon_16ms_happy =[]
list_accuracy_uncon_16ms_sad =[]
list_accuracy_uncon_16ms_neutral =[]

list_accuracy_uncon_25ms_happy =[]
list_accuracy_uncon_25ms_sad =[]
list_accuracy_uncon_25ms_neutral=[]

list_accuracy_con_141ms_happy =[]
list_accuracy_con_141ms_sad =[]
list_accuracy_con_141ms_neutral =[]


acc_happy_list = []
acc_sad_list = []
acc_neutral_list = []
acc_8ms_list = []
acc_16ms_list = []
acc_25ms_list = []
acc_141ms_list = []

accuracy_list = []
participant_ID_list = []
number_corr_ans_list = []
reaction_number_list =[]
```

```python
for i in participant:
    #read in csv for every participant
    participantcsv = pd.read_csv(i, sep =";") #\t , ; ""
    
    #insert index of trials for study part
    length = len(participantcsv)-3
    participantcsv.insert(loc=0, column='trialnumber', value=np.arange(-3,length))
    
    #Delete 3 example trials
    participantcsv=participantcsv.drop([0,1,2])
    
    
    #get correct trials
    def correctness(participantcsv):
        if participantcsv['primeEmotion'] == "sad" and participantcsv['key_resp.keys']== "7":
            return 1
        if participantcsv['primeEmotion'] == "neutral" and participantcsv['key_resp.keys']== "8":
            return 1
        if participantcsv['primeEmotion'] == "happy" and participantcsv['key_resp.keys']== "9":
            return 1
        else:
            return 0
    participantcsv['correct'] = participantcsv.apply(correctness, axis=1)   
    
    #Just get values smaller than 1.5
    #participantcsv = participantcsv[(participantcsv['key_resp.rt'] <= 1.5)] 
    #Drop trials without reaction
    participantcsv = participantcsv.dropna(subset=['key_resp.rt'])

    corr_answers = participantcsv['correct']
    reaction_time = participantcsv['key_resp.rt']

    happy = participantcsv[(participantcsv['primeEmotion'] =='happy')]
    sad = participantcsv[(participantcsv['primeEmotion'] =='sad')]
    neutral = participantcsv[(participantcsv['primeEmotion'] =='neutral')]
    
    uncon_8ms = participantcsv[(participantcsv['PrimerTime120'] == 1)] 
    uncon_16ms = participantcsv[(participantcsv['PrimerTime120'] == 2)] 
    uncon_25ms = participantcsv[(participantcsv['PrimerTime120'] == 3)] 
    con_141ms = participantcsv[(participantcsv['PrimerTime120'] == 18)] 
    
    #Get 12 conditions
    uncon_8ms_happy = uncon_8ms[(uncon_8ms['primeEmotion'] == 'happy')]
    uncon_8ms_sad = uncon_8ms[(uncon_8ms['primeEmotion'] == 'sad')]
    uncon_8ms_neutral = uncon_8ms[(uncon_8ms['primeEmotion'] == 'neutral')]
    
    uncon_16ms_happy = uncon_16ms[(uncon_16ms['primeEmotion'] == 'happy')]
    uncon_16ms_sad = uncon_16ms[(uncon_16ms['primeEmotion'] == 'sad')]
    uncon_16ms_neutral = uncon_16ms[(uncon_16ms['primeEmotion'] == 'neutral')]
    
    uncon_25ms_happy = uncon_25ms[(uncon_25ms['primeEmotion'] == 'happy')]
    uncon_25ms_sad = uncon_25ms[(uncon_25ms['primeEmotion'] == 'sad')]
    uncon_25ms_neutral = uncon_25ms[(uncon_25ms['primeEmotion'] == 'neutral')]
    
    con_141ms_happy = con_141ms[(con_141ms['primeEmotion'] == 'happy')]
    con_141ms_sad = con_141ms[(con_141ms['primeEmotion'] == 'sad')]
    con_141ms_neutral = con_141ms[(con_141ms['primeEmotion'] == 'neutral')]
    
    
    #Excecute functions and get one value for each participant
    val_8ms_happy = accuracy_uncon_8ms_happy(participantcsv,i)
    val_8ms_sad = accuracy_uncon_8ms_sad(participantcsv,i)
    val_8ms_neutral = accuracy_uncon_8ms_neutral(participantcsv,i)
    
    val_16ms_happy = accuracy_uncon_16ms_happy(participantcsv,i)
    val_16ms_sad = accuracy_uncon_16ms_sad(participantcsv,i)
    val_16ms_neutral = accuracy_uncon_16ms_neutral(participantcsv,i)
    
    val_25ms_happy = accuracy_uncon_25ms_happy(participantcsv,i)
    val_25ms_sad = accuracy_uncon_25ms_sad(participantcsv,i)
    val_25ms_neutral = accuracy_uncon_25ms_neutral(participantcsv,i)
    
    val_141ms_happy = accuracy_con_141ms_happy(participantcsv,i)
    val_141ms_sad = accuracy_con_141ms_sad(participantcsv,i)
    val_141ms_neutral = accuracy_con_141ms_neutral(participantcsv,i)
    
    accuracy_for_happy = acc_happy(participantcsv,i)
    accuracy_for_sad = acc_sad(participantcsv,i)
    accuracy_for_neutral = acc_neutral(participantcsv,i)
    accuracy_for_8ms = acc_8ms(participantcsv,i)
    accuracy_for_16ms = acc_16ms(participantcsv,i)
    accuracy_for_25ms = acc_25ms(participantcsv,i)
    accuracy_for_141ms = acc_141ms(participantcsv,i)
    
    participant_ID = name(participantcsv,i)
    number_corr_ans = numbercorrans(participantcsv,i)
    reaction_number = reactionnumber(participantcsv,i)
    accuracy = correct(participantcsv,i)
    
    #append values to append individual value to one list
    list_accuracy_uncon_8ms_happy.append(val_8ms_happy)
    list_accuracy_uncon_8ms_sad.append(val_8ms_sad)
    list_accuracy_uncon_8ms_neutral.append(val_8ms_neutral) 

    list_accuracy_uncon_16ms_happy.append(val_16ms_happy)
    list_accuracy_uncon_16ms_sad.append(val_16ms_sad)
    list_accuracy_uncon_16ms_neutral.append(val_16ms_neutral)

    list_accuracy_uncon_25ms_happy.append(val_25ms_happy)
    list_accuracy_uncon_25ms_sad.append(val_25ms_sad) 
    list_accuracy_uncon_25ms_neutral.append(val_25ms_neutral)

    list_accuracy_con_141ms_happy.append(val_141ms_happy)
    list_accuracy_con_141ms_sad.append(val_141ms_sad) 
    list_accuracy_con_141ms_neutral.append(val_141ms_neutral) 
    
    acc_happy_list.append(accuracy_for_happy)
    acc_sad_list.append(accuracy_for_sad)
    acc_neutral_list.append(accuracy_for_neutral)
    acc_8ms_list.append(accuracy_for_8ms)
    acc_16ms_list.append(accuracy_for_16ms)
    acc_25ms_list.append(accuracy_for_25ms)
    acc_141ms_list.append(accuracy_for_141ms)
    
    participant_ID_list.append(participant_ID)
    number_corr_ans_list.append(number_corr_ans)
    reaction_number_list.append(reaction_number)
    accuracy_list.append(accuracy)
    
    study_number = 'study2'
    
    #Create dataframe 
    accuracy_data = {'name': participant_ID_list,
                     'number of reactions': reaction_number_list,
                     'number of correct answers': number_corr_ans_list,
                     'total trials': accuracy_list,
                     
                     'happy':acc_happy_list,
                     'sad':acc_sad_list,
                     'neutral':acc_neutral_list,
                     '8ms':acc_8ms_list,
                     '16ms':acc_16ms_list,
                     '25ms':acc_25ms_list,
                     '141ms':acc_141ms_list,
                     
                     '8ms happy':list_accuracy_uncon_8ms_happy,
                     '8ms sad':list_accuracy_uncon_8ms_sad,
                     '8ms neutral':list_accuracy_uncon_8ms_neutral,
                     '16ms happy':list_accuracy_uncon_16ms_happy,
                     '16ms sad':list_accuracy_uncon_16ms_sad,
                     '16ms neutral':list_accuracy_uncon_16ms_neutral,
                     '25ms happy':list_accuracy_uncon_25ms_happy,
                     '25ms sad':list_accuracy_uncon_25ms_sad,
                     '25ms neutral':list_accuracy_uncon_25ms_neutral,
                     '141ms happy':list_accuracy_con_141ms_happy,
                     '141ms sad':list_accuracy_con_141ms_sad,
                     '141ms neutral':list_accuracy_con_141ms_neutral,
                    'study_number':study_number}
    
    mean_accuracy = pd.DataFrame(data=accuracy_data).round(4)
```
