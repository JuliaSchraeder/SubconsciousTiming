```python
#import modules 
import pandas as pd
import ast, os
import numpy as np
import csv
import fileinput
import glob
import seaborn as sns
import statistics
```

```python
# get name of participant to reconnect mean data with participant 
def name(participantcsv,vp_number):
    ID_1 =  participantcsv['participant'].tolist()
    ID = ID_1[0]
    ID = ID[-3:]

    return ID 
```

```python
#How often was a key pressed?
def ratingnumber(participantcsv, vp_number):
    rating = participantcsv['rating'].dropna()
    reaction_number = len(rating)                          # number of reactions without missing values
    
    return reaction_number  
```

```python
#read in csv data generated by PsychoPy
participant = glob.glob('.../data_behav_2/*.csv') 
len(participant)
```

```python
list_rating_unconscious =[]
list_rating_conscious=[]

list_rating_happy =[]
list_rating_sad=[]
list_rating_neutral =[]

list_rating_unconscious_happy=[]
list_rating_unconscious_sad =[]
list_rating_unconscious_neutral=[]

list_rating_conscious_happy=[]
list_rating_conscious_sad =[]
list_rating_conscious_neutral =[]

participant_ID_list = []

list_rating_number =[]
```

```python
for i in participant:
    #read in csv for every participant
    participantcsv = pd.read_csv(i, sep =";")
    
    #Delete 3 example trials
    participantcsv=participantcsv.drop([0,1,2])
    
    #Drop trials without reaction
    participantcsv = participantcsv.dropna(subset=['key_resp_Rating.rt'])
    
    #all keypresses are saved as list like this: ['7'] but we just want the 7:
    participantcsv['key_resp_Rating.keys'] = participantcsv['key_resp_Rating.keys'].str[2:3] 
    participantcsv['key_resp_Rating.rt'] = participantcsv['key_resp_Rating.rt'].str[1:8]
    
    #string to numeric 
    participantcsv['key_resp_Rating.rt']=pd.to_numeric(participantcsv['key_resp_Rating.rt'],errors ='coerce')
    participantcsv['key_resp_Rating.keys']=pd.to_numeric(participantcsv['key_resp_Rating.keys'],errors ='coerce')
    rating_time = participantcsv['key_resp_Rating.rt']
    
    #get correct trials
    def correctness(participantcsv):
        if participantcsv['primeEmotion'] == "sad" and participantcsv['key_resp.keys']== "7":
            return 1
        if participantcsv['primeEmotion'] == "neutral" and participantcsv['key_resp.keys']== "8":
            return 1
        if participantcsv['primeEmotion'] == "happy" and participantcsv['key_resp.keys']== "9":
            return 1
        else:
            return 0
    participantcsv['correct'] = participantcsv.apply(correctness, axis=1) 
    corr_answers = participantcsv['correct']
    
    #get rating trials, generate factor for rating response. 1= "dont know", 2 = "unsure", 3="sure!"
    def rating_response(participantcsv):
        number = participantcsv['key_resp_Rating.keys']
        if (number == 7.0) | (number == 7):
            number = 3
        if (number == 8.0) | (number == 8):
            number = 2
        if (number == 9.0) | (number == 9):
            number = 1
        return number
    
    participantcsv['rating'] = participantcsv.apply(rating_response, axis=1)
    rating = participantcsv['rating']
    


    def conditions(participantcsv):  
        if participantcsv['PrimerTime120'] == 18: 
            return "conscious"
        else: 
            return "unconscious"
        
    participantcsv['Conditions'] = participantcsv.apply(conditions, axis=1)
    

    happy = participantcsv[(participantcsv['primeEmotion'] =='happy')]
    sad = participantcsv[(participantcsv['primeEmotion'] =='sad')]
    neutral = participantcsv[(participantcsv['primeEmotion'] =='neutral')]

    unconscious = participantcsv[(participantcsv['Conditions'] == 'unconscious')] 
    conscious = participantcsv[(participantcsv['Conditions'] == 'conscious')] 
    
    
    unconscious_happy = unconscious[(unconscious['primeEmotion'] =='happy')]
    unconscious_sad = unconscious[(unconscious['primeEmotion'] =='sad')]
    unconscious_neutral = unconscious[(unconscious['primeEmotion'] =='neutral')]
    
    conscious_happy = conscious[(conscious['primeEmotion'] =='happy')]
    conscious_sad = conscious[(conscious['primeEmotion'] =='sad')]
    conscious_neutral = conscious[(conscious['primeEmotion'] =='neutral')]

    
    
    #Excecute functions and get one value for each participant
    val_rating_unconscious = rating_unconscious(participantcsv,i)
    val_rating_conscious = rating_conscious(participantcsv,i)
    
    val_rating_happy = rating_happy(participantcsv,i)
    val_rating_sad = rating_sad(participantcsv,i)
    val_rating_neutral = rating_neutral(participantcsv,i)                
                                   
    val_rating_unconscious_happy = rating_unconscious_happy(participantcsv,i)
    val_rating_unconscious_sad = rating_unconscious_sad(participantcsv,i)
    val_rating_unconscious_neutral = rating_unconscious_neutral(participantcsv,i)
    
    val_rating_conscious_happy = rating_conscious_happy(participantcsv,i)
    val_rating_conscious_sad = rating_conscious_sad(participantcsv,i)
    val_rating_conscious_neutral = rating_conscious_neutral(participantcsv,i)
    
    participant_ID = name(participantcsv,i)
    val_rating_number= ratingnumber(participantcsv,i)
    
    #append values to append individual value to one list                             
    list_rating_unconscious.append(val_rating_unconscious)
    list_rating_conscious.append(val_rating_conscious)
    
    list_rating_happy.append(val_rating_happy)
    list_rating_sad.append(val_rating_sad)
    list_rating_neutral.append(val_rating_neutral)
    
    list_rating_unconscious_happy.append(val_rating_unconscious_happy)
    list_rating_unconscious_sad.append(val_rating_unconscious_sad)
    list_rating_unconscious_neutral.append(val_rating_unconscious_neutral)
    
    list_rating_conscious_happy.append(val_rating_conscious_happy)
    list_rating_conscious_sad.append(val_rating_conscious_sad)
    list_rating_conscious_neutral.append(val_rating_conscious_neutral) 
    
    participant_ID_list.append(participant_ID)
    list_rating_number.append(val_rating_number)
    
    #Create dataframe 
    rating_data = {'name': participant_ID_list,
                   'number of rating':list_rating_number,
                     
                     'rating happy':list_rating_happy,
                     'rating sad':list_rating_sad,
                     'rating neutral':list_rating_neutral,
                     'rating unconscious':list_rating_unconscious,
                     'rating conscious':list_rating_conscious,
                     
                     'rating unconscious happy':list_rating_unconscious_happy,
                     'rating unconscious sad':list_rating_unconscious_sad,
                     'rating unconscious neutral':list_rating_unconscious_neutral,
                     
                     'rating conscious happy':list_rating_conscious_happy,
                     'rating conscious sad':list_rating_conscious_sad,
                     'rating conscious neutral':list_rating_conscious_neutral}
    rating_df = pd.DataFrame(data=rating_data)
```
