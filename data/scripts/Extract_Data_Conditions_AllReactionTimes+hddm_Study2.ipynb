```python
#read in csv data generated by PsychoPy
participant = glob.glob('.../data_behav_2/*.csv') 
len(participant)
```

```python
for i in participant:
    #read in csv for every participant
    participantcsv = pd.read_csv(i, sep =";") #\t , ; ""
    
    #insert index of trials for study part
    length = len(participantcsv)-3
    participantcsv.insert(loc=0, column='trialnumber', value=np.arange(-3,length))
    
    #Delete 3 example trials
    participantcsv=participantcsv.drop([0,1,2])
    
    
    def conditions(participantcsv):
        if (participantcsv['PrimerTime120'] == 1) and (participantcsv['primeEmotion'] == 'happy'):
            return "8ms_happy"
        elif (participantcsv['PrimerTime120'] == 1) and (participantcsv['primeEmotion'] == 'sad'):
            return "8ms_sad"
        elif (participantcsv['PrimerTime120'] == 1) and (participantcsv['primeEmotion'] == 'neutral'):
            return "8ms_neutral"        
           
        elif (participantcsv['PrimerTime120'] == 2) and (participantcsv['primeEmotion'] == 'happy'):
            return "16ms_happy"
        elif (participantcsv['PrimerTime120'] == 2) and (participantcsv['primeEmotion'] == 'sad'):
            return "16ms_sad"
        elif (participantcsv['PrimerTime120'] == 2) and (participantcsv['primeEmotion'] == 'neutral'):
            return "16ms_neutral"      
        
        elif (participantcsv['PrimerTime120'] == 3) and (participantcsv['primeEmotion'] == 'happy'):
            return "25ms_happy"
        elif (participantcsv['PrimerTime120'] == 3) and (participantcsv['primeEmotion'] == 'sad'):
            return "25ms_sad"
        elif (participantcsv['PrimerTime120'] == 3) and (participantcsv['primeEmotion'] == 'neutral'):
            return "25ms_neutral"           
        
        elif (participantcsv['PrimerTime120'] == 18) and (participantcsv['primeEmotion'] == 'happy'):
            return "141ms_happy"
        elif (participantcsv['PrimerTime120'] == 18) and (participantcsv['primeEmotion'] == 'sad'):
            return "141ms_sad"
        elif (participantcsv['PrimerTime120'] == 18) and (participantcsv['primeEmotion'] == 'neutral'):
            return "141ms_neutral"              
        else:
            return 0
    
    participantcsv['Conditions'] = participantcsv.apply(conditions, axis=1)
        
    #get primer time
    primer_time_started = participantcsv['primerImage.started']
    primer_time_stopped = participantcsv['primerImage.stopped']
    primer_time = primer_time_stopped - primer_time_started

    participantcsv['primer time'] = primer_time
    
    #Just get values smaller than 1.5
    participantcsv = participantcsv[(participantcsv['key_resp.rt'] <= 1.5)] 
    #Drop trials without reaction
    participantcsv = participantcsv.dropna(subset=['key_resp.rt'])

    
    #get correct trials
    def correctness(participantcsv):
        if participantcsv['primeEmotion'] == "sad" and participantcsv['key_resp.keys']== "7":
            return 1
        if participantcsv['primeEmotion'] == "neutral" and participantcsv['key_resp.keys']== "8":
            return 1
        if participantcsv['primeEmotion'] == "happy" and participantcsv['key_resp.keys']== "9":
            return 1
        else:
            return 0
    participantcsv['correct'] = participantcsv.apply(correctness, axis=1)  
    
    corr_answers = participantcsv['correct']
    reaction_time = participantcsv['key_resp.rt']

    uncon_8ms = participantcsv[(participantcsv['PrimerTime120'] == 1)] 
    uncon_16ms = participantcsv[(participantcsv['PrimerTime120'] == 2)] 
    uncon_25ms = participantcsv[(participantcsv['PrimerTime120'] == 3)] 
    con_141ms = participantcsv[(participantcsv['PrimerTime120'] == 18)] 
    
    #Get 12 conditions
    uncon_8ms_happy = uncon_8ms[(uncon_8ms['primeEmotion'] == 'happy')]
    uncon_8ms_sad = uncon_8ms[(uncon_8ms['primeEmotion'] == 'sad')]
    uncon_8ms_neutral = uncon_8ms[(uncon_8ms['primeEmotion'] == 'neutral')]
    
    uncon_16ms_happy = uncon_16ms[(uncon_16ms['primeEmotion'] == 'happy')]
    uncon_16ms_sad = uncon_16ms[(uncon_16ms['primeEmotion'] == 'sad')]
    uncon_16ms_neutral = uncon_16ms[(uncon_16ms['primeEmotion'] == 'neutral')]
    
    uncon_25ms_happy = uncon_25ms[(uncon_25ms['primeEmotion'] == 'happy')]
    uncon_25ms_sad = uncon_25ms[(uncon_25ms['primeEmotion'] == 'sad')]
    uncon_25ms_neutral = uncon_25ms[(uncon_25ms['primeEmotion'] == 'neutral')]
    
    con_141ms_happy = con_141ms[(con_141ms['primeEmotion'] == 'happy')]
    con_141ms_sad = con_141ms[(con_141ms['primeEmotion'] == 'sad')]
    con_141ms_neutral = con_141ms[(con_141ms['primeEmotion'] == 'neutral')]
    
    
    #Excecute functions and get one value for each participant
    reactionTime_uncon_8ms_happy(participantcsv,i)
    reactionTime_uncon_8ms_sad(participantcsv,i)
    reactionTime_uncon_8ms_neutral(participantcsv,i)
    
    reactionTime_uncon_16ms_happy(participantcsv,i)
    reactionTime_uncon_16ms_sad(participantcsv,i)
    reactionTime_uncon_16ms_neutral(participantcsv,i)
    
    reactionTime_uncon_25ms_happy(participantcsv,i)
    reactionTime_uncon_25ms_sad(participantcsv,i)
    reactionTime_uncon_25ms_neutral(participantcsv,i)
    
    reactionTime_con_141ms_happy(participantcsv,i)
    reactionTime_con_141ms_sad(participantcsv,i)
    reactionTime_con_141ms_neutral(participantcsv,i)
    
    name(participantcsv,i)
    numbercorrans(participantcsv,i)
    reactionnumber(participantcsv,i)
    reactiontotal(participantcsv,i)
    
    participant_ID=name(participantcsv,i)
    
    
    #save written dataframe from each participant
    ## get ID of participant for csv files
    #ID_1 = participantcsv['participant'].tolist()
    #ID = ID_1[0]
    #ID = ID[-3:]

    ## create filename
    file_name=str(participant_ID)+"_Reactions"+".csv"
    df_all_reactions['study_number'] = 'study2'
    # save one csv file for each participant with all reactions
   
    df_all_reactions.to_csv(r'.../FirstLevel/'+file_name ,sep =';', index = False)


    
    ##### HDDM #####
    
    #get reaction times
    reaction_time = participantcsv['key_resp.rt']            # reaction time
    #get accuracy
    correctness = participantcsv['correct']            # get accuracy
    #get emotions from primer
    PrimeMaskTime = participantcsv['PrimerTime120']              # extract emotion
    #get emotions from primer
    PrimeEmotion = participantcsv['primeEmotion']             # extract emotion
    #get correct answers
    corr_answers = participantcsv['correct']
    #get block nummber
    block_number = participantcsv['readinBlocks.thisIndex']
    #get trial number
    trial_number = participantcsv['trials.thisN']
    #get key response
    keyResponse = participantcsv['key_resp.keys']
    #get index of trial numbers (real trial number from 1-360)
    trial_index = participantcsv['trialnumber']
    #get conditions
    conditions = participantcsv['Conditions']
    study_number = 'study2'
    #get primer time
    primer_time = participantcsv['primer time']
    
    #additional dataframe to merge reactions for hddm or GGLM analysis
    d_hddm = {'subj_idx': participant_ID,
               'rt': reaction_time,
              'correct': correctness,
               'response': keyResponse,
               'stim': PrimeEmotion, 
               'level': PrimeMaskTime,
               'real_trial_number': trial_index,
               'trial': trial_number,
               'block': block_number,
               'conditions':conditions,
              'study_number' :study_number,
              'primer_time':primer_time}
    
    df_hddm = pd.DataFrame(data=d_hddm)
    
    
    file_name1=str(participant_ID)+"_hddm_study_2_alltrials"+".csv"
    df_hddm.to_csv(r'...t/FirstLevel/'+file_name1 ,sep =';', index = False)
```
