```python
#read in csv data generated by PsychoPy
participant = glob.glob('.../data_behav/*.csv') 

#test if this had work
len(participant)
```

```python
for i in participant:
    #read in csv for every participant
    participantcsv = pd.read_csv(i, sep =",") #\t , ; ""
    
    #insert index of trials for study part
    length = len(participantcsv)-3
    participantcsv.insert(loc=0, column='trialnumber', value=np.arange(-3,length))
    
    #Delete 5 example trials
    participantcsv=participantcsv.drop([2,3,4,5,6])
    
    
    def conditions(participantcsv):
        if (participantcsv['study mask time'] == 1) and (participantcsv['emotionPrimeStudy'] == 'happy'):
            return "8ms_happy"
        elif (participantcsv['study mask time'] == 1) and (participantcsv['emotionPrimeStudy'] == 'sad'):
            return "8ms_sad"
        elif (participantcsv['study mask time'] == 1) and (participantcsv['emotionPrimeStudy'] == 'neutral'):
            return "8ms_neutral"        
           
        elif (participantcsv['study mask time'] == 2) and (participantcsv['emotionPrimeStudy'] == 'happy'):
            return "16ms_happy"
        elif (participantcsv['study mask time'] == 2) and (participantcsv['emotionPrimeStudy'] == 'sad'):
            return "16ms_sad"
        elif (participantcsv['study mask time'] == 2) and (participantcsv['emotionPrimeStudy'] == 'neutral'):
            return "16ms_neutral"      
        
        elif (participantcsv['study mask time'] == 3) and (participantcsv['emotionPrimeStudy'] == 'happy'):
            return "25ms_happy"
        elif (participantcsv['study mask time'] == 3) and (participantcsv['emotionPrimeStudy'] == 'sad'):
            return "25ms_sad"
        elif (participantcsv['study mask time'] == 3) and (participantcsv['emotionPrimeStudy'] == 'neutral'):
            return "25ms_neutral"           
        
        elif (participantcsv['study mask time'] == 17) and (participantcsv['emotionPrimeStudy'] == 'happy'):
            return "141ms_happy"
        elif (participantcsv['study mask time'] == 17) and (participantcsv['emotionPrimeStudy'] == 'sad'):
            return "141ms_sad"
        elif (participantcsv['study mask time'] == 17) and (participantcsv['emotionPrimeStudy'] == 'neutral'):
            return "141ms_neutral"              
        else:
            return 0
    
    participantcsv['Conditions'] = participantcsv.apply(conditions, axis=1)
        
        
         #get correct trials
    def correctness(participantcsv):
        if participantcsv['corrAnsPrimeStudy'] == participantcsv['key_resp_study.keys']:
            return 1
        else:
            return 0
    participantcsv['correct'] = participantcsv.apply(correctness, axis=1)    
        
    ##### ##### ##### ##### ##### ##### ##### ##### ##### 
    #because of bad timing, we select only trials with good timing ####
    #get primer time
    primer_time_started = participantcsv['image_primer_study.started']
    primer_time_stopped = participantcsv['image_primer_study.stopped']
    primer_time = primer_time_stopped - primer_time_started
    #add column with primer time to csvdata
    participantcsv['primer time'] = primer_time
    ##select data with correct timing #jitter between 4ms (+-2ms)
    #participantcsv_8ms = participantcsv[participantcsv['primer time'].between(0.00633, 0.01033)]
    #participantcsv_16ms = participantcsv[participantcsv['primer time'].between(0.01466,0.01866)]
    #participantcsv_25ms = participantcsv[participantcsv['primer time'].between(0.023,0.027)]
    #participantcsv_141ms = participantcsv[participantcsv['primer time'].between(0.1000,0.15000)]
    ##append dataframes to one corrected dataframe
    #participantcsv = participantcsv_8ms.append(participantcsv_16ms).append(participantcsv_25ms).append(participantcsv_141ms) #append csv to first csv  ]
    
    #Just get values smaller than 1.5
    participantcsv = participantcsv[(participantcsv['key_resp_study.rt'] <= 1.5)] 
    #Drop trials without reaction
    participantcsv = participantcsv.dropna(subset=['key_resp_study.rt'])
    ##### ##### ##### ##### ##### ##### ##### ##### ##### 
    #Get stuff
    #get correct answers
    corr_answers = participantcsv['correct']
    reaction_time = participantcsv['key_resp_study.rt']

    uncon_8ms = participantcsv[(participantcsv['study mask time'] == 1)] 
    uncon_16ms = participantcsv[(participantcsv['study mask time'] == 2)] 
    uncon_25ms = participantcsv[(participantcsv['study mask time'] == 3)] 
    con_141ms = participantcsv[(participantcsv['study mask time'] == 17)] 
    
    #Get 12 conditions
    uncon_8ms_happy = uncon_8ms[(uncon_8ms['emotionPrimeStudy'] == 'happy')]
    uncon_8ms_sad = uncon_8ms[(uncon_8ms['emotionPrimeStudy'] == 'sad')]
    uncon_8ms_neutral = uncon_8ms[(uncon_8ms['emotionPrimeStudy'] == 'neutral')]
    
    uncon_16ms_happy = uncon_16ms[(uncon_16ms['emotionPrimeStudy'] == 'happy')]
    uncon_16ms_sad = uncon_16ms[(uncon_16ms['emotionPrimeStudy'] == 'sad')]
    uncon_16ms_neutral = uncon_16ms[(uncon_16ms['emotionPrimeStudy'] == 'neutral')]
    
    uncon_25ms_happy = uncon_25ms[(uncon_25ms['emotionPrimeStudy'] == 'happy')]
    uncon_25ms_sad = uncon_25ms[(uncon_25ms['emotionPrimeStudy'] == 'sad')]
    uncon_25ms_neutral = uncon_25ms[(uncon_25ms['emotionPrimeStudy'] == 'neutral')]
    
    con_141ms_happy = con_141ms[(con_141ms['emotionPrimeStudy'] == 'happy')]
    con_141ms_sad = con_141ms[(con_141ms['emotionPrimeStudy'] == 'sad')]
    con_141ms_neutral = con_141ms[(con_141ms['emotionPrimeStudy'] == 'neutral')]
    
    
    #Excecute functions and get one value for each participant
    reactionTime_uncon_8ms_happy(participantcsv,i)
    reactionTime_uncon_8ms_sad(participantcsv,i)
    reactionTime_uncon_8ms_neutral(participantcsv,i)
    
    reactionTime_uncon_16ms_happy(participantcsv,i)
    reactionTime_uncon_16ms_sad(participantcsv,i)
    reactionTime_uncon_16ms_neutral(participantcsv,i)
    
    reactionTime_uncon_25ms_happy(participantcsv,i)
    reactionTime_uncon_25ms_sad(participantcsv,i)
    reactionTime_uncon_25ms_neutral(participantcsv,i)
    
    reactionTime_con_141ms_happy(participantcsv,i)
    reactionTime_con_141ms_sad(participantcsv,i)
    reactionTime_con_141ms_neutral(participantcsv,i)
    
    name(participantcsv,i)
    numbercorrans(participantcsv,i)
    reactionnumber(participantcsv,i)
    reactiontotal(participantcsv,i)
    
    participant_ID=name(participantcsv,i)
    
    
    #save written dataframe from each participant
    ## get ID of participant for csv files
    #ID_1 = participantcsv['participant'].tolist()
    #ID = ID_1[0]
    #ID = ID[-3:]

    ## create filename
    file_name=str(participant_ID)+"_Reactions"+".csv"
    df_all_reactions['study_number'] = 'study1'
    # save one csv file for each participant with all reactions
    df_all_reactions.to_csv(r'...t/FirstLevel/'+file_name ,sep =';', index = False)


    
    ##### HDDM #####
    
    #get reaction times
    reaction_time = participantcsv['key_resp_study.rt']            # reaction time
    #get accuracy
    correctness = participantcsv['correct']            # get accuracy
    #get emotions from primer
    PrimeMaskTime = participantcsv['study mask time']              # extract emotion
    #get emotions from primer
    PrimeEmotion = participantcsv['emotionPrimeStudy']             # extract emotion
    #get correct answers
    corr_answers = participantcsv['correct']
    #get block nummber
    block_number = participantcsv['block.thisN']
    #get trial number
    trial_number = participantcsv['trials_study.thisN']
    #get key response
    keyResponse = participantcsv['key_resp_study.keys']
    #get index of trial numbers (real trial number from 1-360)
    trial_index = participantcsv['trialnumber']
    #get conditions
    conditions = participantcsv['Conditions']
    study_number = 'study1'
    #get primer time
    primer_time = participantcsv['primer time']
    
    #additional dataframe to merge reactions for hddm or GGLM analysis
    d_hddm = {'subj_idx': participant_ID,
               'rt': reaction_time,
              'correct': correctness,
               'response': keyResponse,
               'stim': PrimeEmotion, 
               'level': PrimeMaskTime,
               'real_trial_number': trial_index,
               'trial': trial_number,
               'block': block_number,
               'conditions':conditions,
             'study_number':study_number,
             'primer_time':primer_time}
    df_hddm = pd.DataFrame(data=d_hddm)
    
    
    

    
    file_name1=str(participant_ID)+"_hddm_study_1_alltrials"+".csv"
    df_hddm.to_csv(r'.../FirstLevel/'+file_name1 ,sep =';', index = False)
```
