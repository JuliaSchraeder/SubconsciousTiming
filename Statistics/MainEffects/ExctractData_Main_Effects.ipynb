  # Functions to extract Accuracy

- accuracy can be checked in column "key_resp_study.corr"

```python
# how many answers where correct?  in %
# need dropna to delete NANs to get correct number of correct answers
def correct(participantcsv, vp_number):
    corr_ans_drop = participantcsv['key_resp_study.corr'].dropna()   # get list of correct answers
    accuracy = sum(corr_ans_drop)/len(corr_ans_drop)* 100            # get percentage of correct answers
    
    return accuracy
```

```python
# how many answere where correct? in total
# need dropna to delete NANs to get correct number of correct answers
def numbercorrans(participantcsv, vp_number):
    corr_ans_drop = participantcsv['key_resp_study.corr'].dropna()  # get list of correct answers                           
    number_corr_ans_nonformatted = sum(corr_ans_drop) 
    number_corr_ans = "{:.0f}".format(number_corr_ans_nonformatted) # format value to not show .0
    
    return number_corr_ans
```

```python
# execute functions for each participant with comma seperated csv files
for i in participant:
    #read in csv for every participant
    participantcsv = pd.read_csv(i, sep =",") #\t , ; ""
    #insert index of trials for study part (exclude example trials with -6)
    length = len(participantcsv)-6
    participantcsv.insert(loc=0, column='index', value=np.arange(-6,length))
    
    ##### ##### ##### ##### ##### ##### ##### ##### ##### 
    #because of bad timing, we select only trials with good timing ####
    #get primer time
    primer_time_started = participantcsv['image_primer_study.started']
    primer_time_stopped = participantcsv['image_primer_study.stopped']
    primer_time = primer_time_stopped - primer_time_started
    #add column with primer time to csvdata
    participantcsv['primer time'] = primer_time
    #select data with correct timing #jitter between 4ms (+-2ms)
    participantcsv_8ms = participantcsv[participantcsv['primer time'].between(0.00633, 0.01033)]
    participantcsv_16ms = participantcsv[participantcsv['primer time'].between(0.01466,0.01866)]
    participantcsv_25ms = participantcsv[participantcsv['primer time'].between(0.023,0.027)]
    participantcsv_141ms = participantcsv[participantcsv['primer time'].between(0.1000,0.15000)]
    #append dataframes to one corrected dataframe
    participantcsv = participantcsv_8ms.append(participantcsv_16ms).append(participantcsv_25ms).append(participantcsv_141ms) #append csv to first csv  ]
    
    #Just get values smaller than 1.5
    participantcsv = participantcsv[(participantcsv['key_resp_study.rt'] <= 1.5)] 
    #Drop trials without reaction
    participantcsv = participantcsv.dropna(subset=['key_resp_study.rt'])

    ##### ##### ##### ##### ##### ##### ##### ##### ##### 
    
    #get reaction times
    reaction_time = participantcsv['key_resp_study.rt']            # reaction time
    #get emotions from primer
    PrimeMaskTime = participantcsv['study mask time']              # extract emotion
    #get emotions from primer
    PrimeEmotion = participantcsv['emotionPrimeStudy']             # extract emotion
    #get correct answers
    corr_answers = participantcsv['key_resp_study.corr']
    #get block nummber
    block_number = participantcsv['Block']
    #get trial number
    trial_number = participantcsv['study trial']
    #get key response
    keyResponse = participantcsv['key_resp_study.keys']
    #get index of trial numbers (real trial number from 1-360)
    trial_index = participantcsv['index']
    
    #define list for reaction times with mask times, used in functions
    d_mask = {'reaction time': reaction_time, 'mask timing': PrimeMaskTime}
    df_mask = pd.DataFrame(data=d_mask) 
    #define list for reaction times and emotions reaction times
    d_emotion = {'reaction time': reaction_time, 'emotion': PrimeEmotion}
    df_emotion = pd.DataFrame(data=d_emotion) 
    #define list for accuracy and emotions 
    d_emo_acc = {'corr_ans': corr_answers, 'emotion': PrimeEmotion}
    df_emo_acc = pd.DataFrame(data=d_emo_acc)
    #define list for accuracy and mask timing
    d_mask_acc = {'corr_ans': corr_answers, 'mask timing': PrimeMaskTime}
    df_mask_acc = pd.DataFrame(data=d_mask_acc)
    
    #execute function for each participant
    reaction_average = reactionaverage(participantcsv,i)
    reaction_number = reactionnumber(participantcsv,i)
    reaction_average_happy = reactionhappy(participantcsv,i)
    reaction_average_sad = reactionsad(participantcsv,i)
    reaction_average_neutral = reactionneutral(participantcsv,i)
    reaction_average_mask_1 = reactionmask_1frame(participantcsv,i)
    reaction_average_mask_2 = reactionmask_2frame(participantcsv,i)
    reaction_average_mask_3 = reactionmask_3frame(participantcsv,i)
    reaction_average_mask_17 = reactionmask_17frame(participantcsv,i)
    accuracy = correct(participantcsv,i)
    number_corr_ans = numbercorrans(participantcsv,i)
    accuracy_for_happy = acc_happy(participantcsv,i)
    accuracy_for_sad = acc_sad(participantcsv,i)
    accuracy_for_neutral = acc_neutral(participantcsv,i)
    accuracy_for_8ms = acc_8ms(participantcsv,i)
    accuracy_for_16ms = acc_16ms(participantcsv,i)
    accuracy_for_25ms = acc_25ms(participantcsv,i)
    accuracy_for_141ms = acc_141ms(participantcsv,i)
    participant_ID = name(participantcsv,i)
    
    # append returned values to the lists
    reaction_average_list.append(reaction_average) 
    reaction_number_list.append(reaction_number)
    reaction_average_happy_list.append(reaction_average_happy) 
    reaction_average_sad_list.append(reaction_average_sad)
    reaction_average_neutral_list.append(reaction_average_neutral)
    reaction_average_mask_1_list.append(reaction_average_mask_1)
    reaction_average_mask_2_list.append(reaction_average_mask_2)
    reaction_average_mask_3_list.append(reaction_average_mask_3)
    reaction_average_mask_17_list.append(reaction_average_mask_17)
    participant_ID_list.append(participant_ID)
    accuracy_list.append(accuracy)
    number_corr_ans_list.append(number_corr_ans)
    acc_happy_list.append(accuracy_for_happy)
    acc_sad_list.append(accuracy_for_sad)
    acc_neutral_list.append(accuracy_for_neutral)
    acc_8ms_list.append(accuracy_for_8ms)
    acc_16ms_list.append(accuracy_for_16ms)
    acc_25ms_list.append(accuracy_for_25ms)
    acc_141ms_list.append(accuracy_for_141ms)
    
    #additional dataframe to merge reactions for hddm or GGLM analysis
    d_merge = {'subj_idx': participant_ID,
               'rt': reaction_time,
               'response': keyResponse,
               'stim': PrimeEmotion, 
               'level': PrimeMaskTime,
               'real_trial_number': trial_index,
               'trial': trial_number,
               'block': block_number}
    df_merged_reactions = pd.DataFrame(data=d_merge)
    
    #write dataframe for every participant and save lists with mean reactions in "first level"
    d1 = {'name': participant_ID_list,
         'mean reaction time': reaction_average_list,
         'mean reaction time happy': reaction_average_happy_list,
         'mean reaction time sad': reaction_average_sad_list,
         'mean reaction time neutral': reaction_average_neutral_list,
         'mean reaction time primer 8ms': reaction_average_mask_1_list,
         'mean reaction time primer 16ms': reaction_average_mask_2_list,
         'mean reaction time primer 25ms': reaction_average_mask_3_list,
         'mean reaction time primer 141ms': reaction_average_mask_17_list}
    mean_reactions = pd.DataFrame(data=d1)
    
    d2 = {'name': participant_ID_list,
         'number of reactions': reaction_number_list,
         'number of correct answers': number_corr_ans_list,
         'accuracy': accuracy_list,
         'accuracy happy':acc_happy_list,
         'accuracy sad':acc_sad_list,
         'accuracy neutral':acc_neutral_list,
         'accuracy 8ms':acc_8ms_list,
         'accuracy 16ms':acc_16ms_list,
         'accuracy 25ms':acc_25ms_list,
         'accuracy 141ms':acc_141ms_list}
    mean_accuracy = pd.DataFrame(data=d2)
    
    #save written dataframe from each participant
    # get ID of participant for csv files
    ID_1 = participantcsv['participant'].tolist()
    ID = ID_1[0]
    # create filename
    file_name=str(ID)+"_Reactions_selected"+".csv"
    file_name1=str(ID)+"_Reactions_dropna_selected"+".csv"
    # save one csv file for each participant with all reactions
    #df_reactions.to_csv(r'C:/Users/juhoffmann/OneDrive - Uniklinik RWTH Aachen/Auswertung/Pilot2/FistLevel/'+file_name ,sep =';', index = False)
    #df_merged_reactions.to_csv(r'C:/Users/juhoffmann/OneDrive - Uniklinik RWTH Aachen/Auswertung/Pilot2/FistLevel/'+file_name1 ,sep =';', index = False)
    
    df_reactions.to_csv(r'C:/Users/juhoffmann/OneDrive - Uniklinik RWTH Aachen/Auswertung/Pilot2/FistLevel/'+file_name ,sep =';', index = False)
    df_merged_reactions.to_csv(r'C:/Users/juhoffmann/OneDrive - Uniklinik RWTH Aachen/Auswertung/Pilot2/FistLevel/'+file_name1 ,sep =';', index = False)
```
