```python
#read in csv data generated by PsychoPy
participant = glob.glob('C:/Users/juhoffmann/OneDrive - Uniklinik RWTH Aachen/Auswertung/Pilot2/data_behav/*.csv') 
#participant = glob.glob('/Users/julia/OneDrive - Uniklinik RWTH Aachen/Auswertung/Pilot2/data_behav/*.csv') 



#test if this had work
#participant
```

```python
#create lists to append values and to store them in the dataframe df with mean reactions
list_reactionTime_uncon_8ms_happy =[]
list_reactionTime_uncon_8ms_sad=[]
list_reactionTime_uncon_8ms_neutral =[]

list_reactionTime_uncon_16ms_happy=[]
list_reactionTime_uncon_16ms_sad =[]
list_reactionTime_uncon_16ms_neutral=[]

list_reactionTime_uncon_25ms_happy=[]
list_reactionTime_uncon_25ms_sad =[]
list_reactionTime_uncon_25ms_neutral=[]

list_reactionTime_con_141ms_happy=[]
list_reactionTime_con_141ms_sad =[]
list_reactionTime_con_141ms_neutral =[]

reactionTime_list = []
participant_ID_list = []
number_corr_ans_list = []
reaction_number_list =[]
```

```python
for i in participant:
    #read in csv for every participant
    participantcsv = pd.read_csv(i, sep =",") #\t , ; ""
    #insert index of trials for study part (exclude example trials with -6)
    length = len(participantcsv)-6
    participantcsv.insert(loc=0, column='index', value=np.arange(-6,length))

        
    ##### ##### ##### ##### ##### ##### ##### ##### ##### 
    #because of bad timing, we select only trials with good timing ####
    #get primer time
    primer_time_started = participantcsv['image_primer_study.started']
    primer_time_stopped = participantcsv['image_primer_study.stopped']
    primer_time = primer_time_stopped - primer_time_started
    #add column with primer time to csvdata
    participantcsv['primer time'] = primer_time
    #select data with correct timing #jitter between 4ms (+-2ms)
    participantcsv_8ms = participantcsv[participantcsv['primer time'].between(0.00633, 0.01033)]
    participantcsv_16ms = participantcsv[participantcsv['primer time'].between(0.01466,0.01866)]
    participantcsv_25ms = participantcsv[participantcsv['primer time'].between(0.023,0.027)]
    participantcsv_141ms = participantcsv[participantcsv['primer time'].between(0.1000,0.15000)]
    #append dataframes to one corrected dataframe
    participantcsv = participantcsv_8ms.append(participantcsv_16ms).append(participantcsv_25ms).append(participantcsv_141ms) #append csv to first csv  ]
    
    #Just get values smaller than 1.5
    participantcsv = participantcsv[(participantcsv['key_resp_study.rt'] <= 1.5)] 
    #Drop trials without reaction
    participantcsv = participantcsv.dropna(subset=['key_resp_study.rt'])
    ##### ##### ##### ##### ##### ##### ##### ##### ##### 
    #Get stuff
    #get correct answers
    corr_answers = participantcsv['key_resp_study.corr']
    reaction_time = participantcsv['key_resp_study.rt']

    uncon_8ms = participantcsv[(participantcsv['study mask time'] == 1)] 
    uncon_16ms = participantcsv[(participantcsv['study mask time'] == 2)] 
    uncon_25ms = participantcsv[(participantcsv['study mask time'] == 3)] 
    con_141ms = participantcsv[(participantcsv['study mask time'] == 17)] 
    
    #Get 12 conditions
    uncon_8ms_happy = uncon_8ms[(uncon_8ms['emotionPrimeStudy'] == 'happy')]
    uncon_8ms_sad = uncon_8ms[(uncon_8ms['emotionPrimeStudy'] == 'sad')]
    uncon_8ms_neutral = uncon_8ms[(uncon_8ms['emotionPrimeStudy'] == 'neutral')]
    
    uncon_16ms_happy = uncon_16ms[(uncon_16ms['emotionPrimeStudy'] == 'happy')]
    uncon_16ms_sad = uncon_16ms[(uncon_16ms['emotionPrimeStudy'] == 'sad')]
    uncon_16ms_neutral = uncon_16ms[(uncon_16ms['emotionPrimeStudy'] == 'neutral')]
    
    uncon_25ms_happy = uncon_25ms[(uncon_25ms['emotionPrimeStudy'] == 'happy')]
    uncon_25ms_sad = uncon_25ms[(uncon_25ms['emotionPrimeStudy'] == 'sad')]
    uncon_25ms_neutral = uncon_25ms[(uncon_25ms['emotionPrimeStudy'] == 'neutral')]
    
    con_141ms_happy = con_141ms[(con_141ms['emotionPrimeStudy'] == 'happy')]
    con_141ms_sad = con_141ms[(con_141ms['emotionPrimeStudy'] == 'sad')]
    con_141ms_neutral = con_141ms[(con_141ms['emotionPrimeStudy'] == 'neutral')]
    
    
    #Excecute functions and get one value for each participant
    val_8ms_happy = reactionTime_uncon_8ms_happy(participantcsv,i)
    val_8ms_sad = reactionTime_uncon_8ms_sad(participantcsv,i)
    val_8ms_neutral = reactionTime_uncon_8ms_neutral(participantcsv,i)
    
    val_16ms_happy = reactionTime_uncon_16ms_happy(participantcsv,i)
    val_16ms_sad = reactionTime_uncon_16ms_sad(participantcsv,i)
    val_16ms_neutral = reactionTime_uncon_16ms_neutral(participantcsv,i)
    
    val_25ms_happy = reactionTime_uncon_25ms_happy(participantcsv,i)
    val_25ms_sad = reactionTime_uncon_25ms_sad(participantcsv,i)
    val_25ms_neutral = reactionTime_uncon_25ms_neutral(participantcsv,i)
    
    val_141ms_happy = reactionTime_con_141ms_happy(participantcsv,i)
    val_141ms_sad = reactionTime_con_141ms_sad(participantcsv,i)
    val_141ms_neutral = reactionTime_con_141ms_neutral(participantcsv,i)
    
    participant_ID = name(participantcsv,i)
    number_corr_ans = numbercorrans(participantcsv,i)
    reaction_number = reactionnumber(participantcsv,i)
    reaction_total = reactiontotal(participantcsv,i)
    
    #append values to append individual value to one list
    list_reactionTime_uncon_8ms_happy.append(val_8ms_happy)
    list_reactionTime_uncon_8ms_sad.append(val_8ms_sad)
    list_reactionTime_uncon_8ms_neutral.append(val_8ms_neutral) 

    list_reactionTime_uncon_16ms_happy.append(val_16ms_happy)
    list_reactionTime_uncon_16ms_sad.append(val_16ms_sad)
    list_reactionTime_uncon_16ms_neutral.append(val_16ms_neutral)

    list_reactionTime_uncon_25ms_happy.append(val_25ms_happy)
    list_reactionTime_uncon_25ms_sad.append(val_25ms_sad) 
    list_reactionTime_uncon_25ms_neutral.append(val_25ms_neutral)

    list_reactionTime_con_141ms_happy.append(val_141ms_happy)
    list_reactionTime_con_141ms_sad.append(val_141ms_sad) 
    list_reactionTime_con_141ms_neutral.append(val_141ms_neutral) 
    
    participant_ID_list.append(participant_ID)
    number_corr_ans_list.append(number_corr_ans)
    reaction_number_list.append(reaction_number)
    reactionTime_list.append(reaction_total)
    
    
    
    #Create dataframe 
    reactionTime_data = {'name': participant_ID_list,
                     'number of reactions': reaction_number_list,
                     'number of correct answers': number_corr_ans_list,
                     'total trials':reactionTime_list,
                     '8ms happy':list_reactionTime_uncon_8ms_happy,
                     '8ms sad':list_reactionTime_uncon_8ms_sad,
                     '8ms neutral':list_reactionTime_uncon_8ms_neutral,
                     '16ms happy':list_reactionTime_uncon_16ms_happy,
                     '16ms sad':list_reactionTime_uncon_16ms_sad,
                     '16ms neutral':list_reactionTime_uncon_16ms_neutral,
                     '25ms happy':list_reactionTime_uncon_25ms_happy,
                     '25ms sad':list_reactionTime_uncon_25ms_sad,
                     '25ms neutral':list_reactionTime_uncon_25ms_neutral,
                     '141ms happy':list_reactionTime_con_141ms_happy,
                     '141ms sad':list_reactionTime_con_141ms_sad,
                     '141ms neutral':list_reactionTime_con_141ms_neutral}
    mean_reactionTime = pd.DataFrame(data=reactionTime_data)
    
    
    
    

```
